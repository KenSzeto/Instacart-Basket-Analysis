## 04. User Defined Functions

```python
#defining the function

def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] >5) and (row['prices'] <=15):
        return 'Mid-range product'
    elif row['prices'] >15:
        return 'High range product'
    else: return 'Not enough data'
```

```python
df_sub['price_range'] = df_sub.apply(price_label, axis=1)
```

```python
df_sub['price_range'].value_counts(dropna = False)
```

## 05. Deriving Variables with .loc

```python
#Practice with the loc() function
df_sub.loc[df_sub['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_sub.loc[(df_sub['prices'] <= 15) & (df_sub['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df_sub.loc[df_sub['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#loc function on full set

df_ords_prods.loc[df_ords_prods['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

## 06. Task

### Finding Busiest Days

```python
#Finding frequency for busiest days

df_ords_prods['orders_day_of_week'].value_counts(dropna = False)
```

```python
result2 = []

for value in df_ords_prods['orders_day_of_week']:
    if value == 0:
        result2.append('Busiest Days')
    elif value ==1:
        result2.append('Busiest Days')
    elif value == 4:
        result2.append('Least Busy Days')
    elif value == 3:
        result2.append('Least Busy Days')
    else:
        result2.append('Regularly Busy')
```

```python
result2
```

```python
df_ords_prods['busiest_days'] = result2
```

```python
df_ords_prods['busiest_days'].value_counts(dropna = False)
```

```python
df_ords_prods.head()
```

Observations:

The two busiest days of the week are Saturday and Sunday, while the least buys days are Wednesday and Thursday. This shows that people tend to do their grocery shopping over the weekend, while the middle of the week sees less traffic.

### Finding Busiest Hours

```python
df_ords_prods['order_hour_of_day'].value_counts(dropna = False)
```

```python
result3 = []

for value in df_ords_prods['order_hour_of_day']:
    if value < 16 and value > 9:
        result3.append('Most Orders')
    elif value <=6 or value == 23:
        result3.append('Least Orders')
    else:
        result3.append('Average Orders')
```

```python
df_ords_prods['busiest_period_of_day'] = result3
```

```python
df_ords_prods['busiest_period_of_day'].value_counts(dropna = False)
```
